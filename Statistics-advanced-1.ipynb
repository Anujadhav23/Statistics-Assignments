# Theory Questions

# Q.1) What is a random variable in probability theory?
Ans: A random variable is a variable that takes numerical values determined by the outcome of a random experiment. It assigns a numerical value to each possible outcome in the sample space.

# Q.2) What are the types of random variables?
Ans: There are two types of random variables:
1. Discrete Random Variable – takes countable values (e.g., number of heads in 3 coin tosses).
2. Continuous Random Variable – takes uncountable values within a given range (e.g., height of a person).

# Q.3) What is the difference between discrete and continuous distributions?
Ans: 
- Discrete distributions have countable outcomes (like 1, 2, 3…).
- Continuous distributions have infinite outcomes within a given range.
- Example: Discrete → Binomial; Continuous → Normal.

# Q.4) What are probability distribution functions (PDF)?
Ans: A probability distribution function (PDF) describes how probabilities are distributed for a random variable. For discrete variables it's a Probability Mass Function (PMF), and for continuous variables it’s the density function.

# Q.5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
Ans: A PDF gives the probability of the variable taking a specific value (for discrete) or density (for continuous), while the CDF gives the probability that the variable is less than or equal to a certain value (cumulative probability).

# Q.6) What is a discrete uniform distribution?
Ans: A discrete uniform distribution is a distribution where all outcomes are equally likely. Example: rolling a fair six-sided die, each outcome (1–6) has probability 1/6.

# Q.7) What are the key properties of a Bernoulli distribution?
Ans: 
- Only two outcomes: success (1) and failure (0).
- Probability of success = p, failure = 1-p.
- Mean = p, Variance = p(1-p).

# Q.8) What is the binomial distribution, and how is it used in probability?
Ans: The binomial distribution models the number of successes in a fixed number of independent trials, each with the same probability of success. It’s used for problems like number of heads in coin tosses.

# Q.9) What is the Poisson distribution and where is it applied?
Ans: The Poisson distribution models the number of times an event occurs in a fixed interval of time or space. It is used in modeling rare events, like the number of emails received per hour.

# Q.10) What is a continuous uniform distribution?
Ans: A continuous uniform distribution is where every value in a range [a, b] is equally likely. The PDF is constant: f(x) = 1/(b-a) for a ≤ x ≤ b.

# Q.11) What are the characteristics of a normal distribution?
Ans: 
- Bell-shaped and symmetric about the mean.
- Mean = Median = Mode.
- Follows the empirical 68–95–99.7 rule for standard deviations.

# Q.12) What is the standard normal distribution, and why is it important?
Ans: A standard normal distribution is a normal distribution with mean 0 and standard deviation 1. It's important because it simplifies calculations and helps compare values using Z-scores.

# Q.13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Ans: CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the population's distribution. It enables use of normal approximation in statistical inference.

# Q.14) How does the Central Limit Theorem relate to the normal distribution?
Ans: According to CLT, as the sample size increases, the sampling distribution of the sample mean becomes approximately normal, allowing the use of Z-scores and normal probability for analysis.

# Q.15) What is the application of Z statistics in hypothesis testing?
Ans: Z-statistics is used when population variance is known and sample size is large. It helps in testing hypotheses about population means or proportions by comparing the Z value to critical Z scores.

# Q.16) How do you calculate a Z-score, and what does it represent?
Ans: 
Z = (X - μ) / σ  
Where X = value, μ = mean, σ = standard deviation.  
It represents how many standard deviations a value is from the mean.

# Q.17) What are point estimates and interval estimates in statistics?
Ans: 
- Point estimate: a single value estimate of a population parameter (e.g., sample mean).
- Interval estimate: a range within which the parameter likely lies (e.g., confidence interval).

# Q.18) What is the significance of confidence intervals in statistical analysis?
Ans: Confidence intervals indicate the reliability of an estimate. They show a range where the true parameter lies with a certain confidence level (e.g., 95%).

# Q.19) What is the relationship between a Z-score and a confidence interval?
Ans: Z-scores determine the margin of error used to build confidence intervals. For example, a 95% confidence level uses a Z-score of 1.96.

# Q.20) How are Z-scores used to compare different distributions?
Ans: Z-scores standardize different data values, allowing comparison across distributions with different means and standard deviations.

# Q.21) What are the assumptions for applying the Central Limit Theorem?
Ans: 
1. Independent samples
2. Random sampling
3. Large sample size (usually n > 30)
4. Finite variance

# Q.22) What is the concept of expected value in a probability distribution?
Ans: Expected value is the long-run average outcome of a random variable. For discrete X:  
E[X] = ΣxP(x)

# Q.23) How does a probability distribution relate to the expected outcome of a random variable?
Ans: The expected value of a probability distribution gives the average outcome over many repetitions, reflecting the central tendency of the random variable.

# Q.2) Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Discrete Uniform Distribution (e.g., rolling a die)
rv = randint(1, 7)
x = np.arange(1, 7)
pmf = rv.pmf(x)

plt.stem(x, pmf)  # Removed 'use_line_collection=True'
plt.title("PMF of Discrete Uniform Distribution (Die Roll)")
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()
```

# Q.9) Write a Python function to calculate Z-scores from a dataset and plot them

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

dataset = np.random.normal(50, 10, 100)
z_scores = calculate_z_scores(dataset)

plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='purple')
plt.title("Z-Scores Histogram")
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.show()
```

# Q.10) Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
# Exponential distribution (non-normal)
population = np.random.exponential(scale=2, size=10000)

sample_means = []
for i in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='red')
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

# Q.15) Implement hypothesis testing using Z-statistics for a sample dataset

```python
from scipy.stats import norm

# Hypothesis: mean = 50
sample = np.random.normal(52, 5, 50)
sample_mean = np.mean(sample)
population_mean = 50
std_error = 5 / np.sqrt(50)
z = (sample_mean - population_mean) / std_error

p_value = 2 * (1 - norm.cdf(abs(z)))
print("Z-statistic:", z)
print("P-value:", p_value)
```

# Q.16) Create a confidence interval for a dataset using Python and interpret the result

```python
import scipy.stats as stats

sample = np.random.normal(100, 15, 30)
sample_mean = np.mean(sample)
confidence = 0.95
n = len(sample)
std_err = stats.sem(sample)
h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)

lower = sample_mean - h
upper = sample_mean + h
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

# Q.17) Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
data = np.random.normal(200, 20, 50)
mean = np.mean(data)
conf_int = stats.t.interval(0.95, len(data)-1, loc=mean, scale=stats.sem(data))
print("95% Confidence Interval for Mean:", conf_int)
```

# Q.21) Write a Python program to compare the standard deviations of two datasets and visualize the difference

# Q.23) Implement Z-score normalization on a dataset and visualize its transformation

```python
data = np.random.normal(50, 10, 1000)
z_normalized = (data - np.mean(data)) / np.std(data)

plt.hist(z_normalized, bins=30, color='purple', alpha=0.7)
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
